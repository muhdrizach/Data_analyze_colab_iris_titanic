```python
import matplotlib.pyplot as plt
s=sns.get_dataset_names()
print(s)
```

```python
iris=sns.load_dataset('iris')
```

```python
#set style of the plots
sns.set(style="ticks")
```

```python
#pairplot to visualize the relationship in the dataset
sns.pairplot(iris,hue='species')
plt.title("Pairplot of Iris Dataset")
plt.show()
```

```python
#boxplot to visualize distribution of sepal length for each species
plt.figure(figsize=(8,6))
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

```python
#violinplot to visualize distribution of petal width for each species
plt.figure(figsize=(8,6))
sns.violinplot(x='species',y='petal_width',data=iris)
plt.title("Violinplot of Petal Width by Species")
plt.show()
```

```python
#histogram of petal length for each species
plt.figure(figsize=(8,6))
for species in iris['species'].unique():
  subset=iris[iris['species']==species]
  plt.hist(subset['petal_length'],alpha=0.5,label=species)
plt.xlabel("petal length")
plt.ylabel("Frequency")
plt.legend()
plt.title("Histogram of Petal Length by Species")
plt.show()
```

```python
#scatter plot to visualize relationship among sepal lenght and sepal width
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.title("Scatterplot of Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
#load the titanic dataset
titanic=sns.load_dataset('titanic')
print(titanic.head())
```

```python
titanic.info()
```

```python
print(titanic.isnull().sum())
```

```python
#handling missing values
#fill missing values in age with median
median_age=titanic['age'].median()
titanic['age'].fillna(median_age,inplace=True)
```

```python
#fill embark town column with mode
mode_embarktown=titanic['embark_town'].mode()[0]
titanic['embark_town'].fillna(mode_embarktown,inplace=True)
```

```python
#fill embark column with mode
mode_embark=titanic['embarked'].mode()[0]
titanic['embarked'].fillna(mode_embark,inplace=True)
```

```python
#drop the column deck
titanic = titanic.drop('deck', axis=1)
```

```python
print(titanic.info())
```

```python
titanic = titanic.drop(['fare', 'sibsp','parch'], axis=1)
```

```python
print(titanic.info())
```

```python
#save cleaned dataset
titanic=titanic.to_csv('cleaned_titanic.csv')
```

```python
titanic=pd.read_csv('cleaned_titanic.csv')
```

```python
print(titanic.head())
```

```python
titanic.info()
```

```python
plt.figure(figsize=(8,6))
plt.hist(titanic['age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age on Titanic')
plt.show()
```

```python
sex_counts = titanic['sex'].value_counts()
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Passengers by Sex')
plt.show()
```

```python

sns.barplot(x='survived', y='count', data=titanic.groupby('survived').size().reset_index(name='count'))
plt.title('Survival Rate on Titanic')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.histplot(titanic['age'], kde=True)
plt.title('Distribution of Passengers Ages')
plt.xlabel('Age')
plt.ylabel('count')
plt.show()
```

```python
sns.barplot(x='class', y='survived', data=titanic, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=titanic, ci=None)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.barplot(x='embarked', y='survived', data=titanic, ci=None)
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived',data=titanic)
plt.title('Number of Deaths and Survivors by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='class', hue='survived', data=titanic)
plt.title('Number of Deaths and Survivors by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 30, 50, 100], labels=['0-18', '19-30', '31-50', '51+'])
sns.countplot(x='age_group', hue='survived', data=titanic)
plt.title('Number of Deaths and Survivors by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='embarked', hue='survived', data=titanic)
plt.title('Number of Deaths and Survivors by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
survived_sex = titanic.groupby('sex')['survived'].sum()
plt.figure(figsize=(6, 6))
plt.pie(survived_sex, labels=survived_sex.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'pink'])
plt.title('Proportion of Survivors by Sex')
plt.show()
```

```python
survived_class = titanic.groupby('class')['survived'].sum()
plt.figure(figsize=(6, 6))
plt.pie(survived_class, labels=survived_class.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightblue', 'lightcoral'])
plt.title('Proportion of Survivors by Passenger Class')
plt.show()
```
